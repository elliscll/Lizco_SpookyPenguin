```python
import numpy as np 
import pandas as pd 
import os
import datetime as dt
```

```python
data_dir = '/Users/huangha/Desktop/data'
df = pd.read_csv(os.path.join(data_dir, 'Full_Join_LizcoDS_v3.csv'), sep=',', encoding='latin-1')

last_updated = pd.read_csv(os.path.join(data_dir, 'Last Updated Profile By Shop.csv'))
matched = pd.read_csv(os.path.join(data_dir, 'df_final.csv'))
google = pd.read_csv(os.path.join(data_dir, 'All_Shops_Google_Tagged.csv'))
```

```python
matched = matched.drop_duplicates(['company_id'])[['company_id', 'oem']]
```

```python
df = df.merge(last_updated, left_on='company_id', right_on='company_id', how='left')
```

```python
df = df.merge(matched, left_on='company_id', right_on='company_id', how='left')
```

```python
df_google = google[['company_id', 'permanently_closed']].fillna(True)
```

```python
df
```

```python
df = df.drop(['permanently_closed'], axis=1)
df = df.merge(df_google, left_on='company_id', right_on='company_id', how='left')
```

```python
df1 = df.drop(['Unnamed: 0', 'company_id', 'PCD_company_name', 'company_type', 'PCD_ADDRESS', 
               'PCD_CITY', 'PCD_ZIP', 'iso_country_code', 'ABS_ADDRESS', 
               'ABS_CITY', 'ABS_ZIP', 'OBS_ADDRESS', 'OBS_CITY', 'OBS_ZIP', 
               'Integration.Type', 'GOOGLE_ADDRESS', 'GOOGLE_CITY', 
               'GOOGLE_ratio'], axis=1)
```

```python
df1
```

```python
df1['since_last_login'] = dt.datetime.now() - pd.to_datetime(df1['last_login'])
df1['since_last_login'] = df1['since_last_login'] - df1['since_last_login'].min()
df1['since_last_login'] = df1['since_last_login'].dt.total_seconds() / 86400
df1 = df1.drop(['last_login'], axis=1)
```

```python
df1['AAX'] = df1['AAX'].fillna('empty')
df1['Current.Volume'] = df1['Current.Volume'].fillna(-1)
```

```python
def volume_map(x):
    if x > 0:
        return 'positive'
    elif x == 0:
        return 'zero'
    else:
        return 'none'

df1['Current.Volume'] = df1['Current.Volume'].apply(volume_map)
```

```python
df1['is MSO.Name'] = df1['MSO.Name'].isna()
df1['is oem'] = df1['oem'].isna()

df1 = df1.drop(['MSO.Name', 'oem'], axis=1)
```

```python
df1['Suggested Action (Other than Name Change)'] = df1['Suggested Action (Other than Name Change)'].fillna('none').apply(lambda x: True if ((x == 'closed') or (x == 'no body shop')) else False)
```

```python
df1.columns = ['active', 'verified', 'address_type_code', 'is_closed_ABS', 
               'is_AAX', 'current_volume', 'permanently_closed', 'since_last_login',
               'is_mso', 'is_oem']
```

```python
df1
```

```python
df1.to_csv(os.path.join(data_dir, 'feature_engineered.csv'))
```
