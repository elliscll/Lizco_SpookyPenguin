```python
import numpy as np 
import pandas as pd 
import os
from fuzzywuzzy import fuzz
```

```python
data_dir = '/Users/huangha/Desktop/data'
google = pd.read_csv(os.path.join(data_dir, 'All_Shops_Google_Tagged.csv'))
```

```python
google[google['permanently_closed'].isna()]
```

```python
(google['permanently_closed'] == True).mean()
```

```python
(google['permanently_closed'] == False).mean()
```

```python
google[google['company_name'] != google['name']][['company_name', 'name', 'permanently_closed']]
```

```python
google['name'] = google['name'].fillna('')
```

```python
for i in range(len(google)):
    google.loc[i, 'ratio'] = fuzz.partial_ratio(
        google.loc[i, 'company_name'].lower(), google.loc[i, 'name'].lower()
        )
```

```python
google[google['ratio'] < 30][['company_id', 'company_name', 'name', 'permanently_closed', 'ratio']].head(20)
```

```python
google.loc[google['ratio'] < 30, 'permanently_closed'] = True
```

```python
google.to_csv(os.path.join(data_dir, 'google_updated_label.csv'))
```
